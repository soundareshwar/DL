```python
import pandas as pd
import numpy as np
import tensorflow

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Activation
```

```python
import pandas as pd
df = pd.read_excel('sample_data.xlsx')
df
```

```python
from scipy.stats import zscore
df['acceleration'] = zscore(df['acceleration'])
df['collision_impact'] = zscore(df['collision_impact'])
df['image_indicator'] = zscore(df['image_indicator'])

df['severity_of_accident'] = zscore(df['severity_of_accident']) #/df['severity_of_accident'].sum()
df.head()
```

```python
'''df['severity_of_accident'] = df['severity_of_accident']/df['severity_of_accident'].sum()
df['image_indicator'] = df['image_indicator']/df['image_indicator'].sum()

df'''
```

```python
x_train = df.loc[:,'acceleration':'image_indicator'].values
y_train = df.loc[:,'severity_of_accident'].values
```

```python
x_train
```

```python
import tensorflow as tf
model1 = Sequential()
model1.add(Dense(10, input_dim=3, activation='relu'))
model1.add(Dense(1, activation='linear'))
# Compile modelz
optimizer = tf.keras.optimizers.RMSprop(0.0001)
model1.compile(loss='mean_squared_error', optimizer='rmsprop', metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
model1.fit(x_train, y_train, epochs=1000)
```

```python
df.head()
```

```python
print("Model Prediction :",model1.predict([[-0.574768, 0.771702, -0.053368]])[0][0]) # 0.103198
print("Expected :",0.549690)
```

```python
print("Model Prediction :",model1.predict([[ -1.805330, -1.271012, -2.35967 ]])[0][0])
print("Expected :",-1.936366)
```

```python
model2 = Sequential()
model2.add(Dense(10, input_dim=3, activation='relu'))
model2.add(Dense(1, activation='linear'))
# Compile model
model2.compile(loss='mse', optimizer='rmsprop', metrics=['mse', 'mae'])
model2.fit(x_train, y_train, batch_size=3, epochs=1000)
```

```python
print("Model-1 Prediction",model1.predict([[-0.574768, 0.771702, -0.053368]]))
print("Model-2 Prediction",model2.predict([[-0.574768, 0.771702, -0.053368]]))
print("Expected Prediction : ",0.549690)
```
